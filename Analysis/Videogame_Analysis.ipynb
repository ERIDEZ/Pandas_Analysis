## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Unique values counting
Unique_items=Data_df['Item Name'].nunique()

# number of players
Players=Data_df['SN'].count()

# Sum of values
Sum_purch=Data_df['Price'].sum()

# Average purchase price
avr=Sum_purch/Data_df['SN'].count()

# Sum for male players
Sum_males=Data_df[Data_df['Gender']=='Male'].sum()['Price']

# Sum for female players
Sum_female=Data_df[Data_df['Gender']=='Female'].sum()['Price']

# Sum for non disclosed players
Sum_NonDis=Data_df[Data_df['Gender']=='Other / Non-Disclosed'].sum()['Price']

# Count for Other / Non-Disclosed players
Count_NonDis=Data_df[Data_df['Gender']=='Other / Non-Disclosed'].count()['Gender']

# Count for male players
Count_male=Data_df[Data_df['Gender']=='Male'].count()['Gender']

# Count for female players
Count_female=Data_df[Data_df['Gender']=='Female'].count()['Gender']

#print(Count_female+Count_male+Count_NonDis)

Summary_df=pd.DataFrame({
    "Number of Unique Items":Unique_items,
    "Average Purchase Price":avr,
    "Total Number of Purchases":Data_df['Price'].count(),
    "Total Revenue":Sum_purch
}, index=[""])


Summary_df["Average Purchase Price"]=Summary_df["Average Purchase Price"].map("${:,.2f}".format)

Summary_df["Total Revenue"]=Summary_df["Total Revenue"].map("${:,.2f}".format)

Summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Creating a copied DF for eliminating duplicated players

Data_copy=Data_df.copy()
Data_copy=Data_copy.drop_duplicates(subset=['SN'])

sum_price=Data_df.groupby("Gender")["Price"].sum().to_frame(name="Total Purchase Value")
avr_price=round(Data_df.groupby("Gender")["Price"].mean().to_frame(name="Average Purchase Price"),2)
count_price=Data_df.groupby("Gender")["Price"].count().to_frame(name="Purchase Count")

Avg_person=round(Data_df.groupby('Gender')['Price'].sum()/Data_copy.groupby('Gender')['SN'].count(),2).reset_index()
Avg_person=Avg_person.rename(columns={'Gender':'Gender',0:'Avg Total Purchase per Person'}, inplace=False)

Purch_df=sum_price.merge(avr_price, on="Gender")
Purch_df=Purch_df.merge(count_price, on="Gender")
Purch_df=Purch_df.merge(Avg_person, on="Gender")

Purch_df=Purch_df[["Gender","Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

Purch_df.set_index("Gender", inplace=True)

Purch_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Creating the desired operations

Spenders_price=Data_df.groupby('SN')['Price'].sum().to_frame()
Spenders_purch=Data_df.groupby('SN')['Purchase ID'].count().to_frame()
Spenders_average=round(Data_df.groupby('SN')['Price'].sum()/Data_df.groupby('SN')['Purchase ID'].count(),2).reset_index()
```

```python
Spenders_df = Spenders_purch.merge(Spenders_price, on = 'SN')
Spenders_df = Spenders_df.merge(Spenders_average, on = 'SN')

Spenders_df.rename(columns = {'Purchase ID' : 'Purchase Count', 'Price' : 'Total Purchase Value', 0 : 'Average Purchase Price'}, inplace=True)

Spenders_df = Spenders_df[['SN', 'Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]

Spenders_df = Spenders_df.sort_values(by='Total Purchase Value', ascending=False, na_position='first')

Spenders_df.set_index('SN', inplace=True)

Spenders_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a DF for this exercise
Games_df=Data_df.copy()
Games_df=Games_df.drop_duplicates(subset="Item ID", inplace=False)
```

```python
# Extracting essentials

Games_df=Games_df[["Item ID", "Item Name"]]

Games_df
```

```python
# visualizing information required
AverPrice=Data_df.groupby("Item Name")['Price'].mean()
AverPrice
```

```python
# visualizing information required
CountGame=Data_df.groupby("Item Name")['Item ID'].count()

# Creating an already grouped operation
AvgTotalPurc=(AverPrice*CountGame).reset_index()

AvgTotalPurc=AvgTotalPurc.rename(columns={"Item Name":"Item Name", 0:"Total Purchase Value"})

# Visualizing
AvgTotalPurc
```

```python
# Creating first Df for merging
CountGame=round(Data_df.groupby("Item Name")['Item ID'].count().to_frame(),2)
CountGame=CountGame.rename(columns={'Item Name':'Item Name',"Item ID":'Purchase Count'}, inplace=False)

CountGame
```

```python
# Merging of first DF
Games_df=Games_df.merge(CountGame, on="Item Name")
Games_df
```

```python
# Creating and merging second DF
AverPrice=round(Data_df.groupby("Item Name")['Price'].mean().to_frame(),2)
AverPrice=AverPrice.rename(columns={'Item Name':'Item Name',0:'Item Price'}, inplace=False)

Games_df=Games_df.merge(AverPrice, on="Item Name")
Games_df
```

```python
# Merging last DF
Games_df=Games_df.merge(AvgTotalPurc, on="Item Name")
Games_df
```

```python
# Sorting
Games_df = Games_df.sort_values(by='Purchase Count', ascending=False, na_position='first')
Games_df.head(5)
```
